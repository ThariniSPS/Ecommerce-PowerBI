```python

SELECT 
    caldate, 
    eventname,  
    catgroup, 
    priceperticket,
    SUM(qtysold) AS qtysold
FROM event
INNER JOIN category USING(catid)
INNER JOIN date USING(dateid)
INNER JOIN sales USING(eventid)
INNER JOIN listing USING(eventid)
GROUP BY caldate, eventname, catgroup, priceperticket
ORDER BY qtysold DESC
LIMIT 100
```

```python
SELECT 
    o.order_date AS caldate, 
    oi.quantity AS qtysold, 
    c.category_name AS catgroup, 
    oi.list_price AS priceperticket, 
    p.product_name AS eventname
FROM 
    sales.order_items oi
JOIN 
    sales.orders o ON oi.order_id = o.order_id
JOIN 
    production.products p ON oi.product_id = p.product_id
JOIN 
    production.categories c ON p.category_id = c.category_id
WHERE 
    oi.quantity IS NOT NULL 
    AND oi.list_price IS NOT NULL;
```

```python
# Import libraries
import pandas as pd
import plotly.express as px

# Sample data for ticket_prices DataFrame with more data points for clarity
data = {
    "caldate": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"],
    "qtysold": [150, 200, 250, 180, 220],
    "catgroup": ["Concert", "Show", "Concert", "Show", "Concert"],
    "priceperticket": [50, 75, 100, 80, 120],
    "eventname": ["Concert A", "Show B", "Concert C", "Show D", "Concert E"]
}

# Create DataFrame
ticket_prices = pd.DataFrame(data)

# Convert 'caldate' to datetime format for proper plotting
ticket_prices["caldate"] = pd.to_datetime(ticket_prices["caldate"])

# Create scatter plot
fig = px.scatter(
    ticket_prices,
    x="caldate",
    y="qtysold",
    color="catgroup",
    size="priceperticket",
    hover_data=["eventname"],
)

# Update layout with titles and styling
fig.update_layout(
    title="Biggest Concerts and Shows over Time<br><sup>By Quantity Sold and Price of Tickets</sup>",
    title_x=0.5,
    xaxis_title="Date",
    yaxis_title="Quantity Sold",
    legend_title="Category Group"
)

# Show plot
fig.show()
```
